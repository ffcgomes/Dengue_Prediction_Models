```python
import pandas as pd
d=pd.read_csv("./Sobral/Sobral_infectados_Sem_2007.csv", sep=",")
df = pd.DataFrame(data=d)
print(len(df))
```

```python
def inserir_linha(idx, df, df_inserir):
    dfA = df.iloc[:idx, ]
    dfB = df.iloc[idx:, ]
    df = dfA.append(df_inserir).append(dfB).reset_index(drop = True)
    return df
```

```python

j=0
n=1
d_ano=0
while n<=626:
    i=1
    j=j+1
    #print(row['Populacao'])
    while i<7:
        d_inserido = d = {'Dia': [i+1],
        'y':df.iloc[j-1,1]}
        df_inserido = pd.DataFrame(data = d_inserido)
        df = inserir_linha(j, df, df_inserido)

        i=i+1
        j=j+1
    n=n+1
    #print(i,j,n,d_ano)
    
df.to_csv('Fortaleza_Diario_IBGE_Diario.csv')
```
